## Data Cleaning Step 4 : Combining Columns 

Combining arrival date, arrival month and arrival year into 1 single column named arrival_date

```python
# Creating a new 'arrival_date' column by combining 'arrival_date_year', 'arrival_date_month', and 'arrival_date_day_of_month' columns
# Convert 'arrival_date_year' and 'arrival_date_day_of_month' to strings and concatenate them with 'arrival_date_month' using the format 'Year-Month-Day'
# Use pd.to_datetime to convert this concatenated string into a datetime object
# The format '%Y-%B-%d' specifies the year in four digits, the full month name, and the day in two digits
# Setting errors='coerce' ensures that any invalid parsing will result in NaT (Not a Time) instead of throwing an error
data['arrival_date'] = pd.to_datetime(
    data['arrival_date_year'].astype(str) + '-' + 
    data['arrival_date_month'] + '-' + 
    data['arrival_date_day_of_month'].astype(str), 
    format='%Y-%B-%d', errors='coerce'
)
```

```python
# Retrieving and displaying the list of all column names in the dataset
data.columns
```

```python
# Dropping the old arrival date columns
data.drop(['arrival_date_day_of_month'], axis=1, inplace=True)
```

```python
# Retrieving and displaying the list of all column names in the dataset
data.columns
```

```python
# Calculating and returning the total number of columns in the dataset
len(data.columns)
```

```python
# Creating a new column 'kids' by summing the values of 'babies' and 'children' columns
data['kids'] = data['babies'] + data['children']

# Dropping the original 'babies' and 'children' columns from the dataset
# Setting axis=1 indicates columns, and inplace=True modifies the original DataFrame
data.drop(['babies', 'children'], axis=1, inplace=True)

```

```python
# Re-Calculating and returning the total number of columns in the dataset
len(data.columns)
```

## Data Cleaning Step 5  : Adding Season Column

```python
# Defining a function to map the month to its respective season
def get_season(month):
    # December, January, and February are considered Winter
    if month in [12, 1, 2]:
        return 'Winter'
    # March, April, and May are considered Spring
    elif month in [3, 4, 5]:
        return 'Spring'
    # June, July, and August are considered Summer
    elif month in [6, 7, 8]:
        return 'Summer'
    # September, October, and November are considered Autumn
    else:
        return 'Autumn'

# Extracting the month from the 'arrival_date' column and applying the get_season function to map it to a season
data['season'] = data['arrival_date'].dt.month.apply(get_season)

# Displaying the first five rows of the updated DataFrame to verify the changes
data.head()
```

## Data Cleaning Step 5 : Re-Arranging Columns

Rearranging the columns in a more logical order for readability. Placing key columns like 'hotel', 'arrival_date', 'is_canceled', and others first

```python
# Defining a new order for the columns in the DataFrame for better organization and readability
new_column_order = [
    'hotel', 'arrival_date', 'arrival_date_year', 'arrival_date_month', 'season', 
    'arrival_date_week_number', 'lead_time', 'is_canceled', 'adults', 'kids', 
    'stays_in_weekend_nights', 'stays_in_week_nights', 'meal', 'country', 
    'market_segment', 'distribution_channel', 'is_repeated_guest', 
    'previous_cancellations', 'previous_bookings_not_canceled', 'reserved_room_type', 
    'assigned_room_type', 'booking_changes', 'deposit_type', 'days_in_waiting_list', 
    'customer_type', 'adr', 'required_car_parking_spaces', 'total_of_special_requests', 
    'reservation_status', 'reservation_status_date'
]

# Reordering the columns in the DataFrame according to the specified order
data = data[new_column_order]

# Displaying the first five rows of the DataFrame to verify the new column order
data.head()

```

## Data Cleaning Step 6 : Removing the negative values

As adr i.e., average daily rate for the hotel booking cannot be negative, we are dropping all the rows with adr<0.

```python
# Retrieving and displaying the unique values present in the 'adr' (Average Daily Rate) column of the dataset
data['adr'].unique()
```

```python
# Finding and printing the minimum value in the 'adr' (Average Daily Rate) column
print('Min Value in adr: ', min(data['adr']))
```

```python
# Filtering the dataset to include only rows where the 'adr' (Average Daily Rate) is greater than or equal to 0
data = data[data['adr'] >= 0]

# Calculating and returning the total number of rows in the dataset after filtering
len(data)
```

## Data Cleaning Step 7 : Removing Inconsistent Rows

Removing all rows where adults are none and kids are none. As this kind of data has positive adr values along with different types of reseravtion status. When no one is staying in the room then it doesn't make any sense it keep those rows.

```python
# Creating a filter to identify rows where both 'adults' and 'kids' are zero
filtered_data = (data.adults == 0) & (data.kids == 0)

# Displaying the rows in the dataset where both 'adults' and 'kids' are zero
data[filtered_data]
```

```python
# Dropping the rows where both 'adults' and 'kids' are zero by using the negation of the filter
data = data[~filtered_data]

# Calculating and returning the total number of rows in the dataset after dropping the specified rows
len(data)
```

## Data Cleaning Step 8 : Removing Entries 

The other case would be tehre are no adults in the room while there are kids alone staying in the room. This is not a feasible situation

```python
# Creating a filter to identify rows where 'adults' is zero and 'kids' is not zero
cleaned_data = (data.adults == 0) & (data.kids != 0)

# Displaying the rows in the dataset where 'adults' is zero but 'kids' is not zero
data[cleaned_data]
```

# Data Preprocessing

```python
# Extract the year from the 'reservation_status_date' column and create a new column 'res_year'
data['res_year'] = data['reservation_status_date'].dt.year

# Extract the month from the 'reservation_status_date' column and create a new column 'res_month'
data['res_month'] = data['reservation_status_date'].dt.month

# Extract the day from the 'reservation_status_date' column and create a new column 'res_day'
data['res_day'] = data['reservation_status_date'].dt.day

# Extract the day from the 'arrival_date' column and create a new column 'arrival_day'
data['arrival_day'] = data['arrival_date'].dt.day
```

```python
# Drop the 'reservation_status_date' and 'arrival_date' columns from the dataset
# 'axis=1' indicates that columns are being dropped
# 'inplace=True' modifies the original DataFrame without returning a copy
data.drop(['reservation_status_date', 'arrival_date'], axis=1, inplace=True)

# Display the first five rows of the modified dataset to verify the changes
data.head()
```

```python
data['res_year'].unique() # getting unique values from res_year variable
```

```python
# Import the LabelEncoder class from sklearn's preprocessing module
from sklearn.preprocessing import LabelEncoder

# Select all columns of type 'object' (i.e., categorical columns)
categorical_columns = data.select_dtypes(include=['object']).columns

# Initialize a LabelEncoder instance to encode categorical variables into numerical values
label_encoder = LabelEncoder()

# Iterate over each categorical column and apply label encoding
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])
    
# Display the modified dataset with encoded categorical variables
data
```

```python
# Map the 'arrival_date_year' column values to numerical values
# 2015 -> 0, 2014 -> 1, 2016 -> 2, 2017 -> 3
data['arrival_date_year'] = data['arrival_date_year'].map({2015:0, 2014:1, 2016:2, 2017:3})

# Map the 'res_year' column values to numerical values
# 2015 -> 0, 2014 -> 1, 2016 -> 2, 2017 -> 3
data['res_year'] = data['res_year'].map({2015:0, 2014:1, 2016:2, 2017:3})

# Display the first 20 rows of the modified dataset to verify the changes
data.head(20)
```

```python
# Create a new DataFrame 'num_df' by dropping the categorical columns from the original dataset
num_df = data.drop(columns=categorical_columns, axis=1)

# Drop the 'is_canceled' column from 'num_df' as it is likely the target variable and not needed for variance calculation
num_df.drop('is_canceled', axis=1, inplace=True)

# Calculate and display the variance for each numerical column in 'num_df'
num_df.var()
```

```python
# Import StandardScaler and MinMaxScaler from sklearn's preprocessing module
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Define the list of high variance features that will be normalized
high_variance_features = ['lead_time', 'adr', 'country', 'days_in_waiting_list', 
                          'arrival_date_week_number', 'arrival_day', 'res_day', 
                          'arrival_date_month', 'res_month']

# Initialize the MinMaxScaler, which scales features to a range between 0 and 1
scaler_minmax = MinMaxScaler()

# Apply Min-Max scaling (normalization) to the high variance features
# The fit_transform method calculates the min and max for scaling and applies the transformation
data[high_variance_features] = scaler_minmax.fit_transform(data[high_variance_features])

# Print the first few rows of the normalized high variance features to verify the transformation
print(data[high_variance_features].head())
```

## Learning Curves  

```python
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt

# Function to plot learning curves
def plot_learning_curve(estimator, title, X, y, axes=None, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(10, 6))  # Set the figure size for the plot
    
    if ylim is not None:
        plt.ylim(*ylim)  # Set y-axis limits if provided
    
    plt.title(title)  # Set the title of the plot
    plt.xlabel("Training examples")  # Label the x-axis as "Training examples"
    plt.ylabel("Accuracy")  # Label the y-axis as "Accuracy"
    
    # Calculate learning curve data for the given estimator
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring='accuracy'
    )
    
    # Calculate the mean and standard deviation for training and cross-validation scores
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    # Plot the learning curves with grid lines
    plt.grid()

    # Plot the training scores with a filled area representing the standard deviation
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")

    # Plot the cross-validation (testing) scores with a filled area for the standard deviation
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")

    # Display the legend for the training and cross-validation scores
    plt.legend(loc="best")
    return plt

# Plot learning curves for each model in the models dictionary
for model_name, model in models.items():
    # Generate the learning curve for the current model and display it
    plot_learning_curve(model, f'Learning Curve for {model_name}', X_train, y_train, cv=5)
    plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import LinearSVR
from sklearn.neural_network import MLPRegressor
import numpy as np

# Define the parameter grids for each model with reduced combinations to save time
param_distributions = {
    "Linear Regression": {},  # Linear Regression does not have hyperparameters to tune
    
    "Random Forest Regression": {
        'n_estimators': [100, 200],  # Number of trees in the forest
        'max_depth': [None, 10, 20],  # Maximum depth of the tree
        'min_samples_split': [2, 5],  # Minimum number of samples required to split a node
        'min_samples_leaf': [1, 2],  # Minimum number of samples required to be a leaf node
    },
    
    "Support Vector Regression (SVR)": {
        'C': [0.1, 1, 10],  # Regularization parameter
        'epsilon': [0.1, 0.2],  # Epsilon-tube within which no penalty is associated
        'max_iter': [1000, 5000]  # Maximum number of iterations
    },
    
    "Gradient Boosting": {
        'n_estimators': [100, 200],  # Number of boosting iterations
        'learning_rate': [0.01, 0.1],  # Shrinks the contribution of each tree
        'max_depth': [3, 5],  # Maximum depth of the tree
        'min_samples_split': [2, 5],  # Minimum number of samples required to split a node
        'min_samples_leaf': [1, 2]  # Minimum number of samples required to be a leaf node
    },
    
    "Neural Network (MLP)": {
        'hidden_layer_sizes': [(100,), (100, 50)],  # Different architectures for hidden layers
        'activation': ['relu', 'tanh'],  # Activation functions for the hidden layers
        'alpha': [0.0001, 0.001],  # L2 regularization term
        'learning_rate_init': [0.001, 0.0005],  # Initial learning rate
        'max_iter': [1000]  # Maximum number of iterations
    }
}

# Dictionary of regression models
models = {
    "Linear Regression": LinearRegression(),  # Linear Regression model
    "Random Forest Regression": RandomForestRegressor(random_state=42),  # Random Forest Regressor
    "Support Vector Regression (SVR)": LinearSVR(max_iter=10000, random_state=42),  # Support Vector Regressor
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),  # Gradient Boosting Regressor
    "Neural Network (MLP)": MLPRegressor(random_state=42)  # Multi-layer Perceptron Regressor
}

# Train and tune each model using RandomizedSearchCV where applicable
best_models = {}
for model_name, model in models.items():
    if param_distributions[model_name]:  # Check if the model has hyperparameters to tune
        print(f"Tuning {model_name}...")
        random_search = RandomizedSearchCV(
            model, param_distributions[model_name], cv=3, n_iter=5, scoring='r2', n_jobs=1, verbose=2
        )
        random_search.fit(X_train, y_train)  # Fit the RandomizedSearchCV to the training data
        best_models[model_name] = random_search.best_estimator_  # Store the best estimator found
        print(f"Best parameters for {model_name}: {random_search.best_params_}")
    else:
        # For models with no hyperparameters to tune (e.g., Linear Regression)
        model.fit(X_train, y_train)  # Train the model
        best_models[model_name] = model  # Store the trained model

# Evaluate the best models on the test set
results = []
for model_name, best_model in best_models.items():
    y_pred = best_model.predict(X_test)  # Make predictions on the test set
    mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
    mae = mean_absolute_error(y_test, y_pred)  # Mean Absolute Error
    rmse = np.sqrt(mse)  # Root Mean Squared Error
    r2 = r2_score(y_test, y_pred)  # R-squared score
    results.append({
        'Model': model_name,
        'Mean Squared Error (MSE)': mse,
        'Mean Absolute Error (MAE)': mae,
        'Root Mean Squared Error (RMSE)': rmse,
        'R-squared (R2)': r2
    })

# Convert the results into a pandas DataFrame for easier viewing
results_df = pd.DataFrame(results)
results_df  # Display the results DataFrame
```

```python
from sklearn.model_selection import learning_curve

# Function to plot the learning curve
def plot_learning_curve(model, model_name, X_train, y_train):
    plt.figure(figsize=(8, 6))  # Set the figure size for the plot
    
    # Generate learning curve data: train_sizes, train_scores, and validation_scores
    train_sizes, train_scores, validation_scores = learning_curve(
        model, X_train, y_train, cv=5, scoring='r2', n_jobs=-1,  # Perform 5-fold cross-validation, parallelized
        train_sizes=np.linspace(0.1, 1.0, 10)  # Use 10 points on the curve ranging from 10% to 100% of the training set
    )
    
    # Calculate the mean and standard deviation of training scores
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    
    # Calculate the mean and standard deviation of validation (cross-validation) scores
    validation_mean = np.mean(validation_scores, axis=1)
    validation_std = np.std(validation_scores, axis=1)

    # Plot the training scores with error bands representing the standard deviation
    plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training score')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='blue', alpha=0.2)
    
    # Plot the cross-validation scores with error bands representing the standard deviation
    plt.plot(train_sizes, validation_mean, color='green', marker='o', markersize=5, label='Cross-validation score')
    plt.fill_between(train_sizes, validation_mean - validation_std, validation_mean + validation_std, color='green', alpha=0.2)
    
    # Set the title and labels for the plot
    plt.title(f'Learning Curve for {model_name}')
    plt.xlabel('Training Set Size')  # Label the x-axis
    plt.ylabel('R-squared')  # Label the y-axis with R-squared, as that's the metric being used
    
    # Display a legend to differentiate between the training and validation curves
    plt.legend(loc='best')
    
    # Display grid lines on the plot for better readability
    plt.grid(True)
    
    # Show the plot
    plt.show()

# Dictionary to hold the models
# regression_models = {...}  # This is your existing dictionary of models

# Plot learning curves for all models in the regression_models dictionary
for model_name, model in best_models.items():
    print(f"Generating learning curve for {model_name}")  # Indicate which model is being processed
    plot_learning_curve(model, model_name, X_train, y_train)  # Call the function to generate the learning curve
```
